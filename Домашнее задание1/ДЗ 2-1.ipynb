# Задание 1

Необходимо вывести названия треков, которые были приобретены компаниями, которые обслуживаются сотрудниками, нанятыми после 1 сентября 2002 года. 
    

```python
%%sql
select tracks.name as track_name
from tracks, invoice_items, invoices, customers, employees
where tracks.trackid=invoice_items.trackid and
invoice_items.invoiceid=invoices.invoiceid and
invoices.customerid=customers.customerid and
customers.supportrepid=employees.employeeid and
hiredate > '2002-09-1' and customers.company <> 'None'
```

## Задание 2
Вывести общую стоимость и продолжительность каждого плейлиста.

```python
%%sql
select playlists.name, Sum(UnitPrice) as total_cost, Sum(milliseconds) as total_length
from playlist_track, tracks, playlists
where playlist_track.trackid=tracks.trackid and
playlist_track.playlistid=playlists.playlistid
group by playlists.name
```

## Задание 3
Для каждого артиста определить все жанры песен, которые у него есть и количество песен по каждому жанру.


```python
%%sql 
select artists.name, genres.name as genres, count (tracks.name) as number_of_tracks
from artists, albums, tracks, genres
where artists.artistid=albums.artistid and
albums.albumid=tracks.albumid and
tracks.genreid=genres.genreid
group by artists.name, genres.name
order by artists.name
    
```

Для каждого артиста найти жанр, по которому он заработал наибольшее количество денег, а также вывести соотношение общей суммы, заработанной артистом по всем другим жанрам к общей сумме заработанных денег.

```python
%%sql
select artists.name, genres.name as hightest_income_genre, ((totalsum-Max(mun))/totalsum) as percent_of_other_income
from artists, albums, tracks, genres, (select artists.name as artname, genres.name as genrename, sum(UnitPrice) as mun
    from artists, albums, tracks, genres
    where artists.artistid=albums.artistid and
    albums.albumid=tracks.albumid and
    tracks.genreid=genres.genreid
    group by artists.name, genres.name) as tt, 
(select artists.name as port, Sum (Unitprice) as totalsum
    from artists, albums, tracks, genres
    where artists.artistid=albums.artistid and
    albums.albumid=tracks.albumid and
    tracks.genreid=genres.genreid
    group by artists.name ) as zz
where artists.artistid=albums.artistid and
albums.albumid=tracks.albumid and
tracks.genreid=genres.genreid and
artists.name=artname and genres.name=genrename and
artists.name=port
group by artists.name
order by artists.name
```
